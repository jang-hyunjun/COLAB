```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from transformers import BertModel, BertTokenizer
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("multinli.train.ko.tsv" ,sep='\t', on_bad_lines='skip')
```

```python
sentences = df['sentence1'].tolist() + df['sentence2'].tolist()
# NaN 값 처리 및 문자열 변환
sentences = [str(s) for s in sentences if str(s).strip() != ""]
sentences
```

```python
class BERTEmbedder:
    def __init__(self, device='cuda'):
        self.device = torch.device(device if torch.cuda.is_available() else 'cpu')
        self.tokenizer = BertTokenizer.from_pretrained('klue/bert-base')
        self.model = BertModel.from_pretrained('klue/bert-base').to(self.device)
        self.model.eval()

    def embed(self, texts, batch_size=32):
        embeddings = []
        for i in tqdm(range(0, len(texts), batch_size)):
            batch = texts[i:i+batch_size]
            inputs = self.tokenizer(
                batch,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=128
            ).to(self.device)

            with torch.no_grad():
                outputs = self.model(**inputs)

            # [CLS] 토큰 + 평균 풀링 결합 (1536차원)
            cls_emb = outputs.last_hidden_state[:, 0, :]
            mean_emb = outputs.last_hidden_state.mean(dim=1)
            combined = torch.cat([cls_emb, mean_emb], dim=1)
            embeddings.append(combined.cpu())

        return torch.cat(embeddings)

embedder = BERTEmbedder()
embeddings = embedder.embed(sentences)  # (num_sentences, 1536)
```

```python
# VQ-VAE 모델 정의
class VQVAE(nn.Module):
    def __init__(self, input_dim=1536, hidden_dim=512, num_codes=256, code_dim=128):
        super().__init__()
        # 인코더 1536->512->256
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, code_dim)
        )
        # 코드북 256->128
        self.codebook = nn.Embedding(num_codes, code_dim)
        # 디코더 128->256->512->1536
        self.decoder = nn.Sequential(
            nn.Linear(code_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim)
        )

    def forward(self, x):
        z_e = self.encoder(x)  # 인코딩 통과시킴 (256차원의 잠재백터 생성)
        distances = torch.cdist(z_e, self.codebook.weight) # 코드북과 잠재벡터간 거리
        indices = torch.argmin(distances, dim=-1)  # 가장 가까운 코드북 인덱스
        z_q = self.codebook(indices)  # 가장 가까운 벡터로 대입해버림
        x_recon = self.decoder(z_q)  # 재구성
        return x_recon, indices

model = VQVAE().to(embedder.device)
```

```python
# 학습
# 데이터 로더
import torch.optim as optim
dataset = TensorDataset(embeddings) # 학습한 임베딩 불러오기
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# 옵티마이저
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# 학습 루프
for epoch in range(10):
    for batch in tqdm(dataloader):
        x = batch[0].to(embedder.device) # (64, 1536) 크기의 텐서
        optimizer.zero_grad() # 그래디언트 초기화


       # 모델 순전파에서 z_e와 z_q도 추출
        z_e = model.encoder(x)  # 인코더 출력
        distances = torch.cdist(z_e, model.codebook.weight)
        indices = torch.argmin(distances, dim=-1)
        z_q = model.codebook(indices)  # 양자화된 벡터
        x_recon = model.decoder(z_q)  # 재구성 출력

        # 손실 계산
        recon_loss = F.mse_loss(x_recon, x)  # 재구성 손실
        codebook_loss = F.mse_loss(z_e.detach(), z_q)  # 코드북 손실
        commit_loss = F.mse_loss(z_e, z_q.detach())  # 약속 손실
        loss = recon_loss + codebook_loss + 0.25 * commit_loss

        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

```python
# 전체 데이터에 대한 코드북 사용 횟수 계산
all_indices = []
model.eval()  # 평가 모드로 전환
with torch.no_grad():
    for batch in tqdm(dataloader):
        x = batch[0].to(embedder.device)
        _, indices = model(x)
        all_indices.append(indices.cpu())

# 모든 인덱스를 하나의 텐서로 결합
all_indices = torch.cat(all_indices)  # (전체 문장 수,)

# 코드북 사용 통계 계산
codebook_usage = torch.bincount(all_indices, minlength=model.codebook.num_embeddings)
used_codes = torch.sum(codebook_usage > 0).item()

print(f"\n[코드북 사용 현황]")
print(f"전체 코드북 수: {model.codebook.num_embeddings}")
print(f"실제 사용된 코드북 수: {used_codes} ({used_codes/model.codebook.num_embeddings*100:.1f}%)")
```
